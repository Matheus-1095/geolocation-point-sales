```python
import pandas as pd
import geocoder
from geopy.geocoders import Nominatim
from geopy.geocoders import GoogleV3
from geopy.exc import GeocoderTimedOut
from geopy.extra.rate_limiter import RateLimiter
```

```python
df = pd.read_excel("C:/Users/matheus.bulhoes/Desktop/Histórico de Cliente/Report Final/Report Histórico de Clientes.xlsx")
```

```python
df["Endereco_Bairro"] = df[['Endereço', 'Cidade', 'CEP']].astype(str).agg(' - '.join, axis=1)
```

```python
df['Endereco_Bairro'].head(1)
```

```python
locator = Nominatim(user_agent="Mozilla/5.0 (iPhone; CPU iPhone OS 5_1 like Mac OS X) AppleWebKit/534.46 (KHTML, like Gecko) Version/5.1 Mobile/9B179 Safari/7534.48.3")
# 1 - conveneint function to delay between geocoding calls
geocode = RateLimiter(locator.geocode, min_delay_seconds=1)
# 2- - create location column
df['location'] = df['Endereco_Bairro'].apply(geocode)
# 3 - create longitude, laatitude and altitude from location column (returns tuple)
df['point'] = df['location'].apply(lambda loc: tuple(loc.point) if loc else None)
# 4 - split point column into latitude, longitude and altitude columns
df[['latitude', 'longitude', 'altitude']] = pd.DataFrame(df['point'].tolist(), index=df.index)
```
